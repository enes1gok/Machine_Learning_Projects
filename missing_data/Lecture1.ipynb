# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Filling Missing Dataset

```python
from sklearn.impute import SimpleImputer

data = pd.read_csv('missingDataset.csv')

imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')

age = data.iloc[:,1:4].values

imputer = imputer.fit(age[:,1:4])
age[:,1:4] = imputer.transform(age[:,1:4])
age
```

# Categorical Variable to Numeric Variable

```python
country1 = data.iloc[:,0:1].values

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

country1 = le.fit_transform(country1)
print(country1)

country2 = data.iloc[:,0:1].values
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

country2 = ohe.fit_transform(country2).toarray()
print(country2)
```

# Combining Columns

```python
result = pd.DataFrame(data = country2, index = range(22), columns = ['fr', 'tr', 'us'])
print(result.T)
result2 = pd.DataFrame(data = age, index = range(22), columns = ['height', 'weight', 'age'])
print(result2.T)

gender = data.iloc[:,-1].values
gender = le.fit_transform(gender)
result3 = pd.DataFrame(data = gender, index = range(22), columns = ['gender'])
print(result3.T)

combinedData = pd.concat([result, result2, result3], axis = 1)
combinedData

s = pd.concat([result, result2], axis = 1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(s, combinedData, test_size=0.33, random_state=17)

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```
